```python
8. Model Building
```

```python
# Split data into train and test sets
X = np.arange(len(monthly_sales)).reshape(-1, 1)
y = monthly_sales.values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Fit ARIMA model
arima_model = ARIMA(y_train, order=(5,1,0))
arima_model_fit = arima_model.fit()

# Fit SARIMA model
sarima_model = SARIMAX(y_train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_model_fit = sarima_model.fit()
```

```python
arima_model_fit.summary()
```

```python
sarima_model_fit.summary()
```

```python
9. Forecasting
```

```python
arima_predictions = arima_model_fit.forecast(steps=len(X_test))
sarima_predictions = sarima_model_fit.forecast(steps=len(X_test))
# Plot predictions
plt.figure(figsize=(10, 6))
plt.plot(X_train, y_train, label='Train')
plt.plot(X_test, y_test, label='Test')
plt.plot(X_test, arima_predictions, label='ARIMA Predictions')
plt.plot(X_test, sarima_predictions, label='SARIMA Predictions')
plt.title('Sales Forecasting with ARIMA and SARIMA')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import tkinter as tk
from tkinter import messagebox
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# Function to generate forecasted sales for a month
def generate_sales_forecast(start_date):
    # Generate a date range for the month
    date_range = pd.date_range(start=start_date, periods=30, freq='D')
    
    # Simulating forecasted sales (You can replace this with your model's prediction)
    forecasted_sales = np.random.randint(100, 1000, size=len(date_range))
    
    # Create a DataFrame for the forecast
    sales_data = pd.DataFrame({'Date': date_range, 'Forecasted Sales': forecasted_sales})
    
    return sales_data

# Function to plot the forecasted sales
def plot_sales_forecast(sales_data):
    # Clear the previous figure
    for widget in canvas_frame.winfo_children():
        widget.destroy()
    
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(sales_data['Date'], sales_data['Forecasted Sales'], marker='o', linestyle='-', color='b')
    
    ax.set_title('Day-to-Day Sales Forecast for the Month')
    ax.set_xlabel('Date')
    ax.set_ylabel('Forecasted Sales')
    ax.grid(True)
    
    # Display the plot in the Tkinter window
    canvas = FigureCanvasTkAgg(fig, master=canvas_frame)
    canvas.draw()
    canvas.get_tk_widget().pack()

# Function to handle the "Generate Forecast" button click
def on_generate_forecast():
    start_date = entry_date.get()
    
    try:
        # Validate the start date
        pd.to_datetime(start_date)
        
        # Generate sales forecast
        sales_data = generate_sales_forecast(start_date)
        
        # Plot the forecast
        plot_sales_forecast(sales_data)
        
    except Exception as e:
        messagebox.showerror("Invalid Date", "Please enter a valid date in the format YYYY-MM-DD")

# Create the main window
window = tk.Tk()
window.title('Sales Forecasting for One Month')
window.geometry('800x600')

# Add an entry field for the start date
label_date = tk.Label(window, text="Enter the Start Date (YYYY-MM-DD):")
label_date.pack(pady=10)

entry_date = tk.Entry(window)
entry_date.pack(pady=10)

# Add a button to generate the forecast
generate_button = tk.Button(window, text="Generate Forecast", command=on_generate_forecast)
generate_button.pack(pady=20)

# Frame to hold the plot
canvas_frame = tk.Frame(window)
canvas_frame.pack(pady=10)

# Run the Tkinter event loop
window.mainloop()
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.graph_objs as go
import base64
import io
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.model_selection import train_test_split

# Initialize the Dash app
app = dash.Dash(__name__)

# Layout of the app
app.layout = html.Div([
    html.H1("Sales Forecasting Dashboard", style={'text-align': 'center'}),
    
    # File Upload Section
    html.Div([
        dcc.Upload(
            id='upload-data',
            children=html.Button('Upload Sales Data (CSV)'),
            multiple=False
        ),
        html.Div(id='output-data-upload'),
    ], style={'padding': '10px'}),
    
    # Dropdown to select forecast model
    html.Div([
        dcc.Dropdown(
            id='model-dropdown',
            options=[
                {'label': 'ARIMA', 'value': 'ARIMA'},
                {'label': 'SARIMA', 'value': 'SARIMA'}
            ],
            value='SARIMA',  # Default value
            style={'width': '50%'}
        ),
    ], style={'padding': '20px'}),
    
    # Graph for monthly sales and forecasted data
    dcc.Graph(id='sales-forecast-graph'),
])

@app.callback(
    [Output('output-data-upload', 'children'),
     Output('sales-forecast-graph', 'figure')],
    [Input('upload-data', 'contents'),
     Input('model-dropdown', 'value')]
)
def update_output(contents, model_type):
    if contents is None:
        return "No file uploaded yet.", {}
    
    try:
        # Decode the uploaded content
        content_type, content_string = contents.split(',')
        decoded = base64.b64decode(content_string)
        
        # Attempt reading CSV with encoding handling
        try:
            sales_df = pd.read_csv(io.StringIO(decoded.decode('utf-8')))
        except UnicodeDecodeError:
            # Fallback for non-utf-8 encoded files
            sales_df = pd.read_csv(io.BytesIO(decoded), encoding='ISO-8859-1')
        
        # Validate columns
        if 'Order Date' not in sales_df.columns or 'Sales' not in sales_df.columns:
            return "Error: CSV must contain 'Order Date' and 'Sales' columns.", {}

        # Process the data
        sales_df['Order Date'] = pd.to_datetime(sales_df['Order Date'], errors='coerce')
        sales_df = sales_df.groupby('Order Date')['Sales'].sum().reset_index()
        sales_df.set_index('Order Date', inplace=True)

        monthly_sales = sales_df['Sales'].resample('ME').sum()

        
        # Fit the selected model
        if model_type == 'ARIMA':
            model = ARIMA(monthly_sales, order=(5, 1, 0))
        else:  # SARIMA
            model = SARIMAX(monthly_sales, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
        
        model_fit = model.fit()
        
        # Forecast for future months
        forecast = model_fit.forecast(steps=12)
        
        # Plot actual and predicted sales
        trace_actual = go.Scatter(
            x=monthly_sales.index, y=monthly_sales.values,
            mode='lines', name='Actual Sales'
        )
        trace_forecast = go.Scatter(
            x=pd.date_range(monthly_sales.index[-1], periods=12, freq='M'),
            y=forecast, mode='lines', name='Forecasted Sales'
        )
        
        figure = {
            'data': [trace_actual, trace_forecast],
            'layout': go.Layout(
                title=f'{model_type} Sales Forecast',
                xaxis={'title': 'Date'},
                yaxis={'title': 'Sales'},
                showlegend=True
            )
        }
        
        return "File uploaded and processed successfully!", figure
    
    except Exception as e:
        return f"Error: {str(e)}", {}


# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
